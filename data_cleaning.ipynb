```python
#import library
import pandas as pd
import numpy as np
```

```python
#read csv file
job_data = pd.read_csv("dfall_clean.csv")
job_data.head()
```

```python
#total rows and columns
job_data.shape
```

```python
job_data.describe()
```

```python
#information about dataset 
job_data.info()
```

```python
#drop the columns

#job_data.drop(columns = ['redirect_url','adref','created_dt','salary_mid', 'location_area', 'category_tag'], inplace = True)
#job_data.head()

job_data.drop(["redirect_url","adref",'created_dt','salary_mid', 'location_area','category_tag'], axis = 1, inplace = True)
job_data.head()
```

```python
#null values
job_data.isnull().sum()
```

```python
#percentage of null values
job_data.isnull().sum()*100/len(job_data)
```

```python
#Rename columns
job_data.rename({
    'title ': 'job_title',
    'company_name': 'company',
    'category_label': 'job_category',
    'location_display' : 'location',
    'salary_min' : 'min_salary',
    'salary_max' : 'max_salary',
    'salary_is_predicted' : 'predicted_salary',
    'created_date' : 'date_posted'
    },axis=1, inplace = True)
job_data.head()
```

```python
#drop index number
job_data.reset_index(drop= True, inplace = True)
job_data.head()
```

```python
#duplicate value 

job_data.loc[job_data.duplicated(keep=False)]
```

```python
#removing duplicates
job_data.drop_duplicates(keep= 'first',inplace = True)
job_data.shape
```

```python
#check columns datatype
job_data.dtypes
```

```python
#convert date_posted to datetime
job_data['date_posted'] = pd.to_datetime(job_data['date_posted'])

#convert min_salary to integer from float and fill null value with 0
job_data['min_salary'] = job_data['min_salary'].fillna(0).astype(int)

#convert max_salary to integer from float




job_data.dtypes
```
